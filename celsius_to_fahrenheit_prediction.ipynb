```python
import numpy as np
import tensorflow as tf
```

```python
# Create two numpy arrays with at least 7 matching data points (e.g., 0°C = 32°F, 100°C = 212°F, etc.).
celsius_train = np.array([-100.0, 0.0, 4.0, 18.0, 27.0, 32.0, 38.0, 44.0, 50.0, 100.0], dtype = float)
Fahrenheit_train = np.array([-148.0, 32.0, 39.2, 64.4, 80.6, 89.6, 100.4, 111.2, 122.0, 212.0], dtype = float)
```

```python
# Build a 1-neuron model just like our "Hello World" example
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 1, input_shape = [1])
])
```

```python
# Train the model using the sgd optimizer and mean_squared_error for 700 epochs.
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=0.1),
    loss = 'mean_squared_error'
    )
```

```python
# 4. Train the network (let it figure out the pattern over 500 tries)
print("Training the model... please wait a few seconds.")
model.fit(celsius_train, Fahrenheit_train, epochs=1500, verbose=0)
print("Training complete!")
```

```python
# 5. Predict the output for an entirely new Celsius value
new_Celsius = 38.0
prediction = model.predict(np.array([[new_Celsius]]))
print(f"When celsius is {new_Celsius}, TensorFlow predicts Fahrenheit is close to: {prediction[0][0]}")
```
