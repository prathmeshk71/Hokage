```python
import numpy as np
import tensorflow as tf
```

```python
# 1. Create the dataset (y = 2x - 1)
x_train = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
y_train = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

```python
# 2. Build the simplest possible neural network (1 layer, 1 neuron)
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])
```

```python
# 3. Compile the model
model.compile(optimizer='sgd', loss='mean_squared_error')
```

```python
# 4. Train the network (let it figure out the pattern over 500 tries)
print("Training the model... please wait a few seconds.")
model.fit(x_train, y_train, epochs=500, verbose=0)
print("Training complete!")
```

```python
# 5. Predict the output for an entirely new X value
new_x = 2.0
prediction = model.predict(np.array([[new_x]]))
print(f"When X is {new_x}, TensorFlow predicts Y is close to: {prediction[0][0]}")
```
